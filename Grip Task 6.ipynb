##Task :Create the Decision Tree classifier and visualize it graphically.


step1 : IMPORTING THE LIBRARIES

```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing Iris dataset from sklearn.datasets package
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
#Creating dataframe for iris
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['Species'] = iris.target
df['Species Name'] = df['Species'].apply(lambda x: iris.target_names[x])
df.head()
```

```python
#visualize data
df.groupby('Species Name').count()['Species'].plot(kind='bar', color='blue')
plt.ylabel('Number of Samples')
```

```python
#Let's create and train the model
X = df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']]
y = df['Species']
```

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(X,y)

print('Decision Tree Classifer Created')
```

#### Visualization of tree

```python
from sklearn import tree
```

```python
fig, ax = plt.subplots(figsize=(50, 40))
out = tree.plot_tree(dtree,
                  feature_names=iris.feature_names,
                  class_names=iris.target_names,
                  rounded=True,
                  filled=True, 
                  rotate=True, ax=ax,
                  )
```
